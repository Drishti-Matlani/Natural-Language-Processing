# 1

```python
import spacy
nlp=spacy.load("en_core_web_sm")
```

```python
    with open('owlcreek.txt', 'r', encoding='utf-8') as f:
        x = f.read()
doc = nlp(x)
```

# 2

```python
print(f"Number of tokens: {len(doc)}")
```

# 3

```python
print(f"Number of sentences: {len(list(doc.sents))}")
```

# 4

```python
s = list(doc.sents)
ss = s[1]
print("\nSecond sentence:")
print(ss.text)
```

# 5

```python
print(f"{'Text tag':<15} | {'POS tag':<10} | {'Dep tag':<15} | {'Lemma tag':<15}")
for token in ss:
    print(f"{token.text:<15} {token.pos_:<10} {token.dep_:<15} {token.lemma_:<15}")
```

# 6

```python
from spacy.matcher import Matcher
```

```python
matcher = Matcher(nlp.vocab)
pattern = [{"LOWER": "swimming vigorously"}]
matcher.add("swimming vigorously", [pattern])
matches = matcher(doc)
print(f"\nNumber of 'swimming vigorously' matches: {len(matches)}")
```

```python
pharse_list = ['swimming vigorously']

pharse_patterns = [nlp(text) for text in pharse_list]

matcher.add('swimming vigorously' , None , *pharse_patterns)

matches = matcher(doc)
```

# 7

```python
for match_id, start, end in matches:
    span = doc[(start - 5):(end + 5)]
    print(span , "\n" , "-"*10)
```
